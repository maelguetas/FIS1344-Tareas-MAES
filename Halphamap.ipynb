```python
#extracts the FITS header geometry information from a set of files and creates an ASCII 
#image metadata table which is used by several of the other programs
montage.mImgtbl(directory=path_ima, images_table='images_maic.tbl')

#generates the FITS header that best describes the output image.
montage.mMakeHdr('images_maic.tbl', 'template_maic.hdr', north_aligned=True) 
```

```python
montage.mosaic(path_ima, path_ima+'/median', header='template_maic.hdr',background_match=True, combine='median')
```

```python
ima = fits.open('/home/pc/Escritorio/AyInv1s2021/MUSE/img/mosaic.fits')
image = ima[0].data
```

```python
plt.figure(1,figsize=(7,7))
plt.imshow(np.arcsinh(image), origin='lower')#, vmin=np.nanmin(np.arcsinh(image)), 
          # vmax=np.nanmax(np.arcsinh(image)))
plt.xlabel(r'$\rm \Delta \alpha \, [pixels]$', fontsize=16)
plt.ylabel(r'$\rm \Delta \delta \, [pixels]$', fontsize=16)
plt.savefig('SIT45_MUSE_Halpha.png')
```
